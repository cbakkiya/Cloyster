```python
import numpy as np
import statsmodels.api as sm 
import statsmodels.tsa.api as smt 
from  sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from statsmodels.stats.outliers_influence import variance_inflation_factor 
```

```python
target_col = 'total_cases'
```

```python
X = india_case.loc[:, india_case.columns != target_col]
y = india_case.loc[:, target_col]
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.30,
                                                    random_state = 42)
```

```python
X_with_const = sm.add_constant(X_train)
model = sm.OLS(y_train, X_with_const)
```

```python
results = model.fit()
```

```python
results.params
```

```python
print(results.summary())
```

# ***Regression Assumptions Clause***

```python
A_test = sm.add_constant(A_test)
```

```python
A_test
```

```python
b_pred = results.predict(A_test)
```

```python
b_pred
```

```python
residual = b_test - b_pred
```

```python
residual
```

***NO Multicolinearity*** - In statistics, multicollinearity is a phenomenon in which one predictor variable in a multiple regression model can be linearly predicted from the others with a substantial degree of accuracy.

```python
vif = [variance_inflation_factor(A_train.values, i) for i in range (A_train.shape[1])]
pd.DataFrame({'vif': vif[0:]}, index=A_train.columns).T
```

***Normaility of Residual*** - Normality is the assumption that the underlying residuals are normally distributed, or approximately so. 

```python
sns.distplot(residual )
```

```python
pplot =  sm.ProbPlot(residual, fit=True)
fig = pplot.qqplot(line='45')
h = plt.title('Fig. qqplot - residuals of OLS fit')
plt.show()
```

```python
np.mean(residual)
```

***Homoscedasticity*** describes a situation in which the error term (that is, the “noise” or random disturbance in the relationship between the independent variables and the dependent variable) is the same across all values of the independent variables.

```python
fig, ax = plt.subplots(figsize=(10,10))
_ = ax.scatter(b_pred, residual)
```

***No Autocorrelation of residuals*** -  refers to the degree of correlation between the values of the same variables across different observations in the data.

```python
acf = smt.graphics.plot_acf(residual, lags=40, alpha=0.05)
acf.show()
```
